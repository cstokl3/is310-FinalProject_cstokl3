```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import matplotlib_inline
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
# Creating list to append tweet data to
tweets_list_java = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('java AND programming since:2012-01-01 until:2022-01-01').get_items()):
    tweets_list_java.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above
tweets_df_java = pd.DataFrame(tweets_list_java, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
tweets_df_java.to_csv('java.csv')
```
